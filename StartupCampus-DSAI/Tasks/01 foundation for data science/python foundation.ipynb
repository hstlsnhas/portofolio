## Soal 5

Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
* Normal: 18.5-24.9
* Overweight: 25.0-29.9
* Obesity: 30.0-34.9
* Underweight: < 18.5


```python
def status(bmi):
  if bmi > 30 :
    return "Obesity"
  elif bmi > 25 :
    return "Overweight"
  elif bmi > 18.5 :
    return "Normal"
  else:
    return "Underweight"

for bmi in df_24["BMI"]:
  print(status(bmi))
```

```python
df_24["BMI Status"] = df_24["BMI"].apply(status)
df_24.head()
```

# Nomor 2

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:\
Modal Investasi (bilangan bulat antara 0-1.000.000.000)\
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)\
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

\
Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:\
Input
> Money Capital : 100000\
> Investment Tenure (years) : 2

Output
> Year 1, your total investment : 125000.0\
> Year 2, your total investment : 156250.0


```python
from tabulate import tabulate

def Investasi(modal, jangka_waktu):
    # Cek apakah modal dan jangka waktu sesuai ketentuan
    if not 0 < modal <= 1000000000:
        print("\033[94mError: Modal investasi harus dalam rentang antara 1 hingga 1.000.000.000\033[0m")
        return
    if not 0 < jangka_waktu <= 25:
        print("\033[94mError: Jangka waktu investasi harus dalam rentang antara 1 hingga 25 tahun\033[0m")
        return

    # Menghitung keuntungan setiap tahun dan menampilkan informasi dalam tabel
    table = []
    total_modal = modal
    for tahun in range(1, jangka_waktu + 1):
        keuntungan_tahunan = total_modal * 0.25
        total_modal += keuntungan_tahunan
        table.append([tahun, modal, keuntungan_tahunan, total_modal])

    # Tampilkan tabel dengan judul berwarna biru
    headers = ["Tahun", "Modal Awal", "Keuntungan", "Total Modal"]
    print("\033[94m" + tabulate(table, headers=headers, tablefmt="pretty") + "\033[0m")

# Input dari pengguna diluar fungsi investasi
modal_investasi = int(input("Masukkan modal investasi (antara 1 hingga 1.000.000.000): "))
jangka_waktu_investasi = int(input("Masukkan jangka waktu investasi (antara 1 hingga 25 tahun): "))

# Memanggil fungsi investasi dengan input pengguna
Investasi(modal_investasi, jangka_waktu_investasi)
```
